```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import Figure
```

```python
#Section 1: Problem 1

hbl = pd.read_csv("https://raw.githubusercontent.com/fedhere/DSPS_FBianco/master/HW5/hbl2017.csv")

def problem1():
  hbl1 = hbl.iloc[4:20,:4]

  Y = hbl1.y.values
  A = np.c_[np.ones(len(hbl1)),hbl1.x.values]
  C = np.identity(len(hbl1)) * (hbl1["σy"].values**2)

  X = (np.linalg.inv(((A.T).dot(np.linalg.inv(C))).dot(A))).dot(((A.T).dot(np.linalg.inv(C))).dot(Y))
  b = X[0]
  m = X[1]
  x0 = np.linspace(-20,320,200)

  plt.figure(figsize=(5,6),dpi=80)
  plt.errorbar(x=hbl1.x, y=hbl1.y, yerr = hbl1["σy"], fmt='ok', capsize=3)
  plt.plot(x0, m*x0 + b, "darkorange", alpha=0.8, lw=1.25)

  plt.xlim(0,300)
  plt.ylim(0,700)
  plt.xlabel("x");
  plt.ylabel("y");
  plt.title("Figure 1");

  ax = plt.gca();
  ax.minorticks_on();
  ax.tick_params(axis='both',which='both',direction='in')
  ax.yaxis.set_ticks_position('both');
  ax.xaxis.set_ticks_position('both');

  uncMat = np.linalg.inv(A.T.dot(np.linalg.inv(C)).dot(A))
  o2m = np.around(uncMat[0][0],2)
  om = np.around(np.sqrt(uncMat[1][1]),2)
  ob = np.around(np.sqrt(uncMat[0][0]),2)
  print("σ^2_m =", o2m)
  print("σ_m = ", om)
  print("σ_b =", ob)
  eq = "y =" +"(" +str(np.around(m,2)) + "±" +str(om) +")"+"$x$" + " + (" +str(np.around(b,2)) + "±" + str(ob) +")"
  ax.text(50,100,eq,fontsize=12)
  
problem1()
```

The uncertainty for $\sigma^2_m$ on the slope of the line is $332.92$

Caption: This is a plot of the data points 5 through 20 for exercise 1 and a linear best-fit line. The equation for the best-fit line is shown under the line. The best-fit line as shown, goes through most of the data points which shows good linear regression.

```python
#Section 1: Problem 3
def problem3():
  hbl1 = hbl.iloc[4:20,:4]

  Y = hbl1.y.values
  A = np.c_[np.ones(len(hbl1)),hbl1.x.values,hbl1.x.values**2]
  C = np.identity(len(hbl1)) * hbl1["σy"].values**2

  X = (np.linalg.inv(((A.T).dot(np.linalg.inv(C))).dot(A))).dot(((A.T).dot(np.linalg.inv(C))).dot(Y))
  b = X[0]
  m = X[1]
  q = X[2]

  x0 = np.linspace(-20,320,200)

  plt.figure(figsize=(5,6),dpi=80)

  plt.errorbar(x=hbl1.x, y=hbl1.y, yerr = hbl1["σy"], fmt='ok', capsize=3)
  plt.plot(x0, q*(x0**2) + m*x0 + b, "darkorange", alpha=0.8, lw=1.25)

  plt.xlim(0,300)
  plt.ylim(0,700)
  plt.xlabel("x");
  plt.ylabel("y");
  plt.title("Figure 3");

  ax = plt.gca();
  ax.minorticks_on();
  ax.tick_params(axis='both',which='both',direction='in')
  ax.yaxis.set_ticks_position('both');
  ax.xaxis.set_ticks_position('both');

  uncMat = np.linalg.inv(A.T.dot(np.linalg.inv(C)).dot(A))
  oq = np.around(np.sqrt(uncMat[2][2]),4)
  o2m = np.around(uncMat[0][0],2)
  om = np.around(np.sqrt(uncMat[1][1]),2)
  ob = np.around(np.sqrt(uncMat[0][0]),2)
  print("σ_q", oq)
  print("σ^2_m =", o2m)
  print("σ_m = ", om)
  print("σ_b =", ob)
  
  eq = "y = (" +str(np.around(q,4)) + "±" +str(oq) +")"+'$x^2$'+" + "+"(" +str(np.around(m,2)) + "±" +str(om) +")"+"$x$" + " + (" +str(np.around(b,2)) + "±" + str(ob) +")"
 
  ax.text(5,50,eq,fontsize=9.8)

  
  
problem3()
```

Caption: This is a plot of the data points 5 through 20 for exercise 3. Instead of a linear regression like in exercise 1, here we do a quadratic regression. The equation for the best-fit line is shown under the line. The best-fit line as shown, goes through most of the data points which shows good quadratic regression.
It looks hard to tell, but it seems to hit less points than the linear best-fit line in exercise 1, or both best-fit lines could be very close.
